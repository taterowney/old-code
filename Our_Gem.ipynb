# Step 1: Show $ d[e^\theta]/d\theta = e^\theta$ 

> Step 1.1: rephrase the limit that computes $ e $

> Step 1.2: Apply the definition of the derivative to $e^x$

> Step 1.3: Perform a change of variables

> Step 1.4: Reduce the result via algebra

> Step 1.5: Observe the result

***

## Step 1.1

Since $ e =  \lim_{n \to \infty} (1+ 1/n)^n $

Compute the same limit with $ m = 1/n $

This yields: $ e =  \lim_{m \to 0} (1+ m)^{1/m} $

***

## Step 1.2

$ \frac{d e^x}{dx} =  \lim_{\delta \to 0} \frac{e^{x+\delta} \ - \  e^x}{\delta}$

$  =  \lim_{\delta \to 0} \frac{e^x e^{\delta} \ - \  e^x}{\delta}$

$  =  \lim_{\delta \to 0} \frac{e^x (e^{\delta} \ - \  1)}{\delta}$

$  =  e^x \lim_{\delta \to 0} \frac{e^{\delta} \ - \  1}{\delta}$

***

## Step 1.3

> Let $\epsilon = e^\delta - 1$  That means:

> $\epsilon + 1 = e^\delta$ 

so 

> $ln(\epsilon + 1) = \delta$

***

## Step 1.4

 so now from above $ e^x \lim_{\delta \to 0} \frac{e^{\delta} \ - \  1}{\delta}$

=  $ e^x \lim_{\epsilon \to 0} \frac{\epsilon + 1 - 1}{ln(\epsilon + 1)}$

=  $ e^x \lim_{\epsilon \to 0} \frac{\epsilon}{ln(\epsilon + 1)}$

=  $ e^x \lim_{\epsilon \to 0} \frac{1}{\frac{1}{\epsilon}ln(\epsilon + 1)}$

=  $ e^x \lim_{\epsilon \to 0} \frac{1}{ ln((\epsilon + 1)^ \frac{1}{\epsilon} )}$

=  $ e^x  \frac{1}{ ln( \lim_{m \to 0}  (m + 1)^ \frac{1}{m} )}$

***

## Step 1.5


And from the new limit for e above....

=  $ e^x  \frac{1}{ ln(e)}$

$ \therefore \dfrac{d e^x}{dx} = e^x $

***

# Taylor Series Expansions

Recall the definition of Taylor Series Expansions:

> $ f(x) = \displaystyle \sum_{n=0}^{\infty} \dfrac{f^{(n)}(0) x^n}{n!} $


So since  $ \dfrac{d e^x}{dx} = e^x $

We have $ e^x = \displaystyle \sum_{n=0}^{\infty} \dfrac{x^n}{n!} $

So $ e^{i \theta} = \displaystyle \sum_{n=0}^{\infty} \dfrac{(i \theta)^n}{n!} $

Thus $ e^{i \theta} = \displaystyle \sum_{n=0}^{\infty} i^n \dfrac{\theta^n}{n!} $
***

Also recall that 
> $ \dfrac{d(sin(x))}{dx} = cos(x) $

> $ \dfrac{d(cos(x))}{dx} = -sin(x) $

> $ \dfrac{d(-sin(x))}{dx} = -cos(x) $

> $ \dfrac{d(-cos(x))}{dx} = sin(x) $

so
> $ \dfrac{d^{(n)}(sin(x))}{dx} = \left\{\begin{array}{lr}
        sin(x), & \text{for } n = 0 \mod 4\\
        cos(x), & \text{for } n = 1 \mod 4\\
        -sin(x),& \text{for } n = 2 \mod 4\\
        -cos(x),& \text{for } n = 3 \mod 4\\
        \end{array}\right\}$

> $ \dfrac{d^{(n)}(cos(x))}{dx} = \left\{\begin{array}{lr}
        cos(x), & \text{for } n = 0 \mod 4\\
        -sin(x),& \text{for } n = 1 \mod 4\\
        -cos(x),& \text{for } n = 2 \mod 4\\
        sin(x), & \text{for } n = 3 \mod 4\\
        \end{array}\right\}$


And from basic trigonometry:
> $ sin(0)= 0 $

> $ cos(0)= 1 $


thus
> $ \dfrac{d^{(n)}(sin(0))}{dx} = \left\{\begin{array}{lr}
        0, & \text{for } n = 0 \mod 4\\
        1, & \text{for } n = 1 \mod 4\\
        0,& \text{for } n = 2 \mod 4\\
        -1,& \text{for } n = 3 \mod 4\\
        \end{array}\right\}$

> $ \dfrac{d^{(n)}(cos(0))}{dx} = \left\{\begin{array}{lr}
        1, & \text{for } n = 0 \mod 4\\
        0,& \text{for } n = 1 \mod 4\\
        -1,& \text{for } n = 2 \mod 4\\
        0, & \text{for } n = 3 \mod 4\\
        \end{array}\right\}$


***

So now we have

$ sin(x) = \sum\limits_{k\ odd}^{\infty} (-1)^{(k-1)/2} \dfrac{x^n}{n!} $

$ cos(x) = \sum\limits_{k\ even}^{\infty} (-1)^{k/2} \dfrac{x^n}{n!} $

***

Consider the Taylor expansion for $ cos(x) + i sin(x)$

$ cos(x) + i\ sin(x) =  \displaystyle \sum_{n=0}^{\infty} i^n \dfrac{\theta^n}{n!}  = e^{ix}$
